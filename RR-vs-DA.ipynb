# WMT16: Comparison of Relative Ranking and Direct Assessment Scores

```python
import pandas
scores = pandas.read_csv("scores.dat")
print scores.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.stats
fig = plt.figure(figsize = (10,15))
for i,pair in  enumerate(scores['pair'].unique()):
    selected = scores[scores['pair'] == pair]
    ax = fig.add_subplot(421 + i)
    ax.set_xlabel("Rank score")
    ax.set_ylabel("Adequacy")
    
    x,y = selected['rr'], selected['adequacy'] 
    ax.scatter(x,y)
    slope, intercept, r_value, _,_ = scipy.stats.linregress(x, y)
    ax.plot(x, x*slope+intercept, color='red')
    ax.set_title("Pair: {} Corr: {}".format(pair,r_value))
plt.tight_layout()
plt.show()
```
